# Dictionaries

## What are Dictionaries?

Dictionaries are an unordered collection of keys and values. 

Dictionaries are written with curly brackets {}, this creates an empty dictionary.

They can be very useful because you can store information as various datatypes, you can nest different data types and you don't need to have the same information for every key (Ex. Storing patient data as a dictionary)

Format: dict = {key: value, key: value}


The keys allow you to access data using immutable (not changeable) objects like strings and numbers. You can't use any mutable datatypes, like a list as a key because it can be modified in place. 

Keys must be unique within the individual dictionary. 

car = {'brand': 'Ford', 'model': 'Mustang', 'color': 'black'}				  
       {   key:   value,  key:     value,     key:    value  }

Values in a dictionary can be ANY PYTHON DATA TYPE, including arrays, other dictionaries, etc.

Ex. Values as lists
animals = {'mammals': ['dog', 'cat', 'bear'], 'fish': ['trout', 'bass', 'salmon']}

Ex. Values as other dictionaries

nested_dict = { 'dictA': {'key_1': 'value_1'},
                'dictB': {'key_2': 'value_2'}}
          
dictA and dictB are the keys of the first dictionary while key_1 and key_2 are the keys of the nested dictionaries

## How are they different than lists? 

1. You access using the keys, not position
2. Dictionaries are unordered, for example you can't access a dictionary by saying dictionary_name[1] 
3. Dictionaries can contain other datatypes, including lists, numpy arrays, other dictionaries, etc 

## How can we see what's in it and what the properties are? 

You can print(dict_name) to print the contents of a dictionary 

In order to see what keys are in a dictionary you use the following: 
dict_name.keys()

In order to see what keys are in a dictionary you use the following: 
dict_name.values()

In order to see what is stored at a specific key, you use:
dict_name[key] -> value

In order to see how many keys you have, you can:
print(len(dict_name))

In order to see how many values you have for a specific key, you can:
print(len(dict_name[key]))

```python
#Examples
animals = {'mammal': ['cat','dog'],'reptile':['lizard','snake']}
print("This is the entire dictionary")
print(animals)
print()

print("These are all the keys in the dictionary")
x = animals.keys()
print(x)
print()

print("These are all the values in the dictionary")
y = animals.values()
print(y)
print()

print("accessing the values of the key mammal")
z = animals['mammal']
print(z)
print()

print("These are all the keys in our dictionary") 
print(len(animals))
print()

print("These are all the values in our dictionary for the key mammal") 
print(len(animals['mammal']))
```

## How do you manipulate or change it? 

To add a key/value pair, you can specify the key name and associated value as follows:
dict['key'] = value 

To change a value for a key, you can access the key and update the value in the same way you add a new key/value pair.
dict['key'] = new_value 
If you are using a list you can do an append command or other data type manipulation 
dict['key'].append("new_value")

To delete a key value pair completely (instead of updating it to a new value / list) you can use del:
del(dic['key'])

```python
#Examples
animals = {'mammal': ['cat','dog'],'reptile':['lizard','snake']}
print("This is the entire dictionary")
print(animals)
print()

#Adding a new key 
animals['marsupial'] = ['playtpus','koala']
print("The dictionary with the new key marsupial")
print(animals)
print()

#updating an exisiting key
animals['marsupial'].append("kangaroo")
print("The dictionary with the updated marsupial key")
print(animals)
print()

#Deleting a key
del(animals['reptile'])
print("The dictionary with the reptile key removed")
print(animals)
```

## Here are some more complicated / flopy dictionary examples

```python
#nested dictionary example
#This is an example of how dictionaries can contain other complex datatypes
#for example, other dictionaries. All of the above methods hold, 
#with slightly different ways of accessing the information 

#create the dictionary 
people = {'dict1': {'name': 'John', 'age': '27', 'sex': 'Male'},
          'dict2': {'name': 'Marie', 'age': '22', 'sex': 'Female'}}
#print the output of the nested dictionary 
print("This is our nested dictionary")
print(people)
print()

#Accessing values from the different levels of dictionaries 
print("All the values from key dict1 in the first dictionary")
print(people['dict1'])
print("Printing the values of the name key in dict1")
print(people['dict1']['name'])
print()

#Changing or adding elements in a nested dictionary 
people['dict3']= {} #creating an empty dictionary for the key 'dict3'
print(people)
print()
#fill the key dict3 with information 
people['dict3']['name'] = 'Luna'
people['dict3']['age'] = '24'
people['dict3']['sex'] = 'Female'
people['dict3']['married'] = 'No'
print(people['dict3'])

#We can delete elements from a nest dictionary 
del people['dict3']['married']
print("deleting the key married in dict3")
print(people['dict3'])
print()
#we can also delete an entire nested dictionary 
del people['dict3']
print("deleting a whole dictionary")
print(people)

```

The next cell just defines variables for us to set up the ET and Recharge correctly
and then read in the head data for the example. This is taken from Self_Check of HWK5
but for only 6 years instead of 16. Feel free to skip to the next section. 

```python
#some code needed to show the example from HW5
import flopy
import numpy as np
import matplotlib as mp
import os
import flopy.utils.binaryfile as bf
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import csv
%matplotlib inline 

nrow = 50 #number of rows
ncol = 50 #number of columns
nlay = 1  #number of layers

dx= 100 #width of grid cells in x diretion 
dy= 100 #width of grid cells in y diretion 

Lx = ncol*dx  #width of domain in x
Ly = nrow*dy #width of domain in y
ztop = 100.    #top elevation 
zbot = 0.     #bottom elevation

dz = (ztop - zbot) / nlay #cell height in z direction

ET_val1 = 0 #m/day
ET_val2 = 1e-3 #m/day
ET_locxy=[[400, 900], [1500, 1600]] #lower left and upper right corners of ET in xy coordinate
extinction_depth = 100

pumping = 0 #m3/day  *This rate was figured out from the Water balance of the ET scenario
well_xy=np.average(ET_locxy, axis=0) #xy location of well in center of ET zone
well_row = np.floor(nrow-(well_xy[1]/dy)) #convert the y location to a row
well_col=np.floor(well_xy[0]/dx) #convert the x location to a column
well_loc = (0,well_row,well_col) #Well loc shoudl be layer, row, column


sp_peryr = 2        # periods for seasonality (two stress periods per year)
sp1_len = 30       # Length of stress period 1 in days
sp2_len = 335      # Length of stress period 2 in days
sim_years = 6       #number of years for transient simulation 
nper = int(sp_peryr*sim_years + 1) # total number of stress periods the +1 is for the steady state start

steady = np.ones(nper, dtype=bool)*False     # steady-state or transient flag for each stress period (boolean array)
steady[0] = True    # initial stress period is steady-state

perlen=np.tile([sp1_len, sp2_len],sim_years)
perlen=np.append(10,perlen) #add a period of length 10 at the start for the steady state simulation

nstp=perlen*1
nstp[0]=1 #for the steady state solution we just want one output

ET_rows=(np.floor(nrow-ET_locxy[1][1]/dy),np.floor(nrow-ET_locxy[0][1]/dy)) #convert the y locations to rows
ET_cols=(np.floor(ET_locxy[0][0]/dx),np.floor(ET_locxy[1][0]/dx)) #convert the y locations to rows
ET_zone = np.zeros((nlay,nrow,ncol))     # define an array of zeroes the size of the model grid
ET_zone[0,int(ET_rows[0]):int(ET_rows[1]), int(ET_cols[0]):int(ET_cols[1])] = 1                 # assign ET rate to the ET zone


#Setup alternating ET values and Recharge
ET_val=np.tile([ET_val1, ET_val2],sim_years)
ET_val=np.append(ET_val1,ET_val) #add a period of length 10 at the start for the steady state simulation

recharge = 1e-4 #m/day
rech_zone = np.zeros((nrow,ncol))               #define an array of zeros of dim (nrow,ncol)
rech_zone=rech_zone+recharge
```

## Dictionaries for MODFLOW inputs
In our scripts we often use dictionaries to set up variables to before we run 
MODFLOW (like Output Control, well locs, Recharge, or ET) but we also use them to organize many of our outputs. 

For example, a dictionary was created for both recharge (R) and Evapotranspiration (ET) where 
the **key** is the stress period number and the **value** is a numpy array for the ET for each grid well. likewise for the recharge (R). 

In this example, there are two stress periods for 6 years of simulation and 1 SS period. 

```python
R = {}                                             # create empty dictionary for recharge fluxes
for p in range(0,nper):                            # loop over stress periods, skipping initial steady state defined above
    R[p]=rech_zone

ET = {}
for p in range(0,nper):    #loop over stress periods
    ET[p]=ET_zone*ET_val[p]
```

## Dictionaries to for MODFLOW Outputs
We also use dictionaries to save outputs in a convenient way. For example, for each day 
in the simulation we can read in the head output file (.hsd) and set the day as the **key**, and a numpy array written as **value**, where each grid cell has an assigned head value. 

Note that the 'sp%s'%(stress_per) portion is a replace method, similar to the .format command.
With 6 years of simulation and 1 SS solution there are 2191 days of head values. 

```python
hds = bf.HeadFile('Example_Heads'+'.hds')                   # reads the binary head file
times = hds.get_times()                               # returns a list of timesteps


#extract binary data from head file
head = {} #create dictionary to store head data at end of each stress period
headobj = flopy.utils.binaryfile.HeadFile('Example_Heads'+'.hds') #get head data as python object

#get data from python objects
for stress_per, time in enumerate(times): #iterate through times at end of each stress period
    head['sp%s'%(stress_per)] = headobj.get_data(totim=time) #append heads to head list for ea stress per
```

We also referenced dictionaries in a loop to create 3D graphs in homework 5.  We pulled out the times with respect to stress periods and associated them with the head values in our dictionary.

```python
#set X, Y, Z variables for 3d plot to be our model domain and head solution
X = np.arange(0,Lx,dx)
Y = np.arange(0,Ly,dy)
X, Y = np.meshgrid(X, Y)
times = [perlen[0],perlen[0]+perlen[1],perlen[0]+perlen[1]+perlen[2]] #extract times at end of each stress period

print (len(times))

#create a figure for every time
for i in range(len(times)):
    #create 3d figure
    fig_3d = plt.figure(figsize=(12,5)) #create a figure instance
    ax = fig_3d.gca(projection='3d') #set an axes with 3d properties
    Z = np.flipud(head['sp%s'%i][0]) #head matrix is flipped to display properly 

    #create surface and labels
    surf = ax.plot_surface(X,Y,Z, cmap = plt.cm.coolwarm, linewidth=0, antialiased=False, label='head') #plot head surface
    fig_3d.colorbar(surf,shrink=0.5,aspect=5).set_label('Head (m)',fontsize=10,fontweight='bold') #set colorbar
    ax.set_xlabel('Lx (m)', fontsize=15, fontweight='bold')
    ax.set_ylabel('Ly (m)', fontsize=15, fontweight='bold')
    ax.set_title('Head Surface: Stress-Period %s'%(i+1), fontsize=15, fontweight='bold')
    plt.show(surf)
```
